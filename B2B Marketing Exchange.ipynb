```python
# b2b_marketing_exchange_scraper.py

import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

def setup_selenium():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Run in headless mode (optional)
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    return driver

def fetch_page_source(url):
    driver = setup_selenium()
    driver.get(url)
    page_source = driver.page_source
    driver.quit()
    return page_source

def parse_main_page(page_source):
    soup = BeautifulSoup(page_source, "html.parser")
    soup = BeautifulSoup(soup.prettify(), "html.parser")

    title = "B2B Marketing Exchange"

    event_dates = soup.find_all("span", attrs={'class': 'elementor-icon-list-text'})[0]
    dates = event_dates.get_text().strip()

    location = soup.find_all("span", attrs={'class': 'elementor-icon-list-text'})[1]
    location = location.get_text().strip()

    description_elements = soup.find_all('span', class_='NormalTextRun SCXW200178519 BCX0')
    description = ' '.join([desc.get_text().strip() for desc in description_elements])

    speakers = "No Details"

    return title, dates, location, description, speakers

def parse_agenda_page():
    url = "https://b2bmarketing.exchange/east/agenda/"
    driver = setup_selenium()
    driver.get(url)

    wait = WebDriverWait(driver, 10)
    try:
        iframe = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'iframe[src*="swapcard.com/widget/event/b2b-marketing-exchange-east-2024"]')))
        driver.switch_to.frame(iframe)
    except:
        pass

    wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'sc-dab8fe09-5')))
    page_source = driver.page_source
    driver.quit()

    soup = BeautifulSoup(page_source, "html.parser")
    agenda_elements = soup.find_all("h3", attrs={"class":"sc-dab8fe09-5 fwiXyw"})
    agenda = ", ".join([ag.get_text() for ag in agenda_elements])

    return agenda

def parse_registration_page():
    url = "https://b2bmarketing.exchange/east/registration/"
    page_source = fetch_page_source(url)
    soup = BeautifulSoup(page_source, "html.parser")

    registration_details = f'Can register through this link. {url}'

    price_elements = soup.find_all("div", attrs={"class": "elementor-price-table__price"})
    prices = ", ".join([price.find("span", attrs={"class": "elementor-price-table__integer-part"}).get_text().strip() for price in price_elements])

    return registration_details, prices

def main():
    main_page_url = "https://b2bmarketing.exchange/"
    main_page_source = fetch_page_source(main_page_url)

    title, dates, location, description, speakers = parse_main_page(main_page_source)
    agenda = parse_agenda_page()
    registration_details, pricing = parse_registration_page()

    categories = "B2B Marketing Summit"
    audience_type = 'All'

    event_details = {
        "Event Name": title,
        "Event Date(s)": dates,
        "Location": location,
        "Website URL": main_page_url,
        "Description": description,
        "Key Speakers": speakers,
        "Agenda": agenda,
        "Registration Details": registration_details,
        "Pricing": pricing,
        "Categories": categories,
        "Audience type": audience_type
    }

    print(event_details)
    df = pd.DataFrame([event_details])
    df.to_csv('B2B_Marketing_Exchange.csv', index=False)

    df_from_csv = pd.read_csv('B2B_Marketing_Exchange.csv')
    print(df_from_csv)

if __name__ == "__main__":
    main()
```
