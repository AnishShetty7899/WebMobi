```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def initialize_webdriver():
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Run in headless mode (optional)
    service = Service(ChromeDriverManager().install())
    return webdriver.Chrome(service=service, options=options)

def get_page_source(url):
    driver = initialize_webdriver()
    driver.get(url)
    page_source = driver.page_source
    driver.quit()
    return page_source

def extract_event_details():
    url = "https://www.sales30conf.com/sales-conference-june2024/?sp_src=upcoming_event"
    page = get_page_source(url)
    Soup1 = BeautifulSoup(page, "html.parser")
    Soup2 = BeautifulSoup(Soup1.prettify(), "html.parser")

    # Title
    title = Soup2.find("h1", attrs={"class": "displayTitle"}).get_text().strip()

    # Date
    date = ''
    dt = Soup2.find_all("div", attrs={"class": "pricingDescription"})
    for i in dt:
        d = i.find("span", attrs={"style": "color:#333333;"}).get_text().strip()
        date = "".join(d)
    split_date = date.split()
    Dates = " ".join(split_date[14:])

    # Audience type
    Aud_Typ = []
    Audience_type1 = Soup2.find_all("div", attrs={"class": "floatRight padding"})
    Audience_type2 = Soup2.find_all("div", attrs={"class": "floatLeft padding"})
    for i in Audience_type1:
        audr = i.find_all("span", attrs={"style": "color:#333333;"})
        for j in audr:
            aur = j.get_text().strip()
            Aud_Typ.append(aur)
    for i in Audience_type2:
        audlf = i.find_all("span", attrs={"style": "color:#333333;"})
        for j in audlf:
            aulf = j.get_text().strip()
            Aud_Typ.append(aulf)
    Aud_type_list = ", ".join(Aud_Typ)

    # Price
    price = Soup2.find("div", attrs={"class": "pricingRate"}).get_text().strip()

    # Location
    loca = "Virtual Event"

    # Description
    Description = "N/A"

    # Speakers
    url1 = "https://www.sales30conf.com/sales-conference-june2024/speakers.html?sp_src=june2024-header-txt-speakers"
    page1 = get_page_source(url1)
    Soup3 = BeautifulSoup(page1, "html.parser")
    Soup4 = BeautifulSoup(Soup3.prettify(), "html.parser")

    s = []
    Speaker = Soup4.find_all("span", attrs={"class": "red_speaker"})
    for i in Speaker:
        speak = i.get_text().strip()
        s.append(speak)
    Speakers = ", ".join(s)

    # Agendas
    url2 = "https://www.sales30conf.com/sales-conference-june2024/agenda.html"
    page2 = get_page_source(url2)
    Soup5 = BeautifulSoup(page2, "html.parser")
    Soup6 = BeautifulSoup(Soup5.prettify(), "html.parser")

    a = []
    Agenda = Soup6.find_all("div", attrs={"class": "AgendaSession"})
    for i in Agenda:
        agenda = i.find_all("span", attrs={"class": "agendaSessionTitle"})
        for j in agenda:
            agnd = j.get_text().strip()
            a.append(agnd)
    Agendas = ", ".join(a)

    # Registration details
    rg_details = 'Can register through this link: https://sales30conference.swoogo.com/june2024/begin?ref=june2024-header-txt-register'

    # Categories
    categories = "B2B Marketing Summit"

    event_details = {
        "Event Name": title,
        "Event Date(s)": Dates,
        "Location": loca,
        "Website URL": url,
        "Description": Description,
        "Key Speakers": Speakers,
        "Agenda": Agendas,
        "Registration Details": rg_details,
        "Pricing": price,
        "Categories": categories,
        "Audience type": Aud_type_list
    }

    return event_details

def save_to_csv(event_details, filename='CRO_Summit.csv'):
    df = pd.DataFrame([event_details])
    df.to_csv(filename, index=False)
    return pd.read_csv(filename)

if __name__ == "__main__":
    event_details = extract_event_details()
    print(event_details)
    df_from_csv = save_to_csv(event_details)
    print(df_from_csv)
```
